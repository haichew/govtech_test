```python
import requests
import json
import pandas as pd
```

```python
response = requests.get('https://api.covid19api.com/dayone/country/singapore').text
```

```python
response
```

```python
response_info = json.loads(response)
response_info
```

```python
response_info[0].keys()
```

```python
df = pd.DataFrame.from_dict(response_info)
df
```

```python
df.info()
```

### Since we do nto need ID, country, countrycode, etc, we only keep the necessary columns which are confirmed, deaths, recovered, active and date 

```python
df_sg = df[['Confirmed', 'Deaths', 'Recovered', 'Active', 'Date']]
df_sg
```

### Next, we convert the Date column to proper datetime format

```python
df_sg['Date'] = pd.to_datetime(df_sg['Date']).dt.strftime('%d/%m/%Y')
df_sg
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=2, cols=2)

fig.add_trace(
    go.Scatter(x=df['Date'], y=df['Confirmed'], mode='lines', name='Confirmed Cases'),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=df['Date'], y=df['Deaths'], mode='lines', name='Deaths'),
    row=1, col=2
)
fig.add_trace(
    go.Scatter(x=df['Date'], y=df['Recovered'], mode='lines', name='Recovered Cases'),
    row=2, col=1
)
fig.add_trace(
    go.Scatter(x=df['Date'], y=df['Active'], mode='lines', name='Active Cases'),
    row=2, col=2
)

fig.update_layout(
    autosize=False,
    width=1200,
    height=800,
    title_text= 'Breakdown of Covid19 Cases in Singapore over time'
)


fig.show()
```
